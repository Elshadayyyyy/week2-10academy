```python
import pandas as pd

df = pd.read_csv('data/raw/reviews_raw.csv')
df.head()
```

```python
df = df.drop_duplicates(subset='review_text')  
df = df.dropna(subset=['review_text'])        
df.to_csv('data/clean/cleaned_reviews.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('data/clean/cleaned_reviews.csv')
df.head()
```

```python
from textblob import TextBlob

df['sentiment_score'] = df['review_text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label'] = df['sentiment_score'].apply(categorize_sentiment)

df[['review_text', 'sentiment_score', 'sentiment_label']].head()
```
