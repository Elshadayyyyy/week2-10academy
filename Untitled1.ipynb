```python
import psycopg2
import pandas as pd
from textblob import TextBlob


conn = psycopg2.connect(
    host="localhost",
    port="5432",
    database="fintech_reviews",
    user="postgres",
    password="1997Elshu."
)

query = """
SELECT 
    r.id,
    r.bank_name,
    r.review_text,
    b.bank_id
FROM reviews r
LEFT JOIN banks b
    ON r.bank_name = b.bank_name;
"""

df = pd.read_sql(query, conn)

def get_sentiment(text):
    if not text or pd.isna(text):
        return 0
    return TextBlob(text).sentiment.polarity

df["sentiment_score"] = df["review_text"].apply(get_sentiment)

def label_sentiment(score):
    if score > 0.1:
        return "Positive"
    elif score < -0.1:
        return "Negative"
    else:
        return "Neutral"

df["sentiment_label"] = df["sentiment_score"].apply(label_sentiment)

print(" Total reviews:", len(df))
print(df["sentiment_label"].value_counts())


bank_summary = df.groupby("bank_name")["sentiment_score"].mean().reset_index()
bank_summary.columns = ["bank_name", "avg_sentiment"]

print("\n Average Sentiment Per Bank:")
print(bank_summary)


df.to_csv("sentiment_results.csv", index=False)
bank_summary.to_csv("bank_summary.csv", index=False)

print("\n Files saved: sentiment_results.csv and bank_summary.csv")
```
